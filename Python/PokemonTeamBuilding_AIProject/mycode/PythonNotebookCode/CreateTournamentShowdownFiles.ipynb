```python
# Imports

import importlib
import sys
import os
sys.path.append(os.path.abspath(".."))

import PythonCode.LoadTeam
import PythonCode.ConvertToShowdown

importlib.reload(PythonCode.LoadTeam)
importlib.reload(PythonCode.ConvertToShowdown)

from UniversalFunctions import *
from PythonCode.LoadTeam import LoadandValidateTeam
from PythonCode.ConvertToShowdown import ConvertTeamToShowdownJSON, GetTournamentTeamDatasets
```

```python
# Get the files of the tournament teams and the directory to put them
datasetFile = "../../data/Teams/limitless_tournament_426_teams_formatted.json"
showdownTeamDir = "../../data/ShowdownFormattedTeams/Limitless426"

regulationPool = LoadJSON("../../data/pokemon_regulation_i.json")
regulationRules = LoadJSON("../../data/regulation_i_rules.json")
items = LoadJSON("../../data/items.json")

dataset = LoadJSON(datasetFile)
teamsData = GetTournamentTeamDatasets(dataset)
```

```python
# Create the showdown files of the tournament teams
tempTeamFile = "../../data/Output/temp_team_validation.json"
for i, entry in enumerate(teamsData, start=1):
    placement = entry.get("placement", i)
    player = entry.get("player", f"player_{i}")
    team = entry["team"]

    WriteJSON(tempTeamFile, team)
    
    result = LoadandValidateTeam(tempTeamFile, regulationPool, regulationRules, items)
    
    showdownTeam = ConvertTeamToShowdownJSON(result["team"], regulationRules["battle_level"])
    
    outputFile = os.path.join(showdownTeamDir,f"placement_{placement}_{player}_showdown.json")
    
    WriteJSON(outputFile, showdownTeam)
    print(f"Result Accepted: {i} -> {outputFile}")
    
if os.path.exists(tempTeamFile):
    os.remove(tempTeamFile)

    
```
